```python
import numpy as np
from pandas import to_datetime as dt
import matplotlib.pyplot as plt
import pandas as pd
data={"Date": dt(["2004-01-03", "2004-02-08", "2004-03-10", "2004-04-25", "2004-05-26", "2004-06-24","2004-07-05", "2004-08-09", "2004-09-05", "2004-10-21", "2004-11-30", "2004-12-21", "2004-12-16"]),
      "Value": [10,20,40,30,60,45,70,100,80,50,10,80,110,]}
df = pd.DataFrame(data)
df.set_index("Date", inplace=True)
plt.figure(figsize=(10, 6))
plt.plot(df.index, df["Value"], marker='o')
plt.xlabel("Data")
plt.ylabel("Value")
plt.title("Time Series with trend")
plt.grid(True)
plt.show()
```
